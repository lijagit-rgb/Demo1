```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```


#Here the model predicts whether a person of certain age , drawing salary of certain amount can purchase A four wheeler SUV of high end or not.  

Dataset created 2 columns : Age & Salary

Import dataset

```python
dataset=pd.read_csv('/content/Social_Network_Ads (4).csv')
x=dataset.iloc[:, :-1].values
y=dataset.iloc[:, -1].values

print(x)
```

split dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=.25, random_state=0)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

Training The Model

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

Predicting the result

```python
print(classifier.predict(sc.transform([[41, 90000]])))
```

prdicting the test results

```python
y_pred=classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test),1)),1))
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test, y_pred)
print(cm)
accurarcy_score=(y_test, y_pred)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

# Set professional theme
sns.set_theme(style="whitegrid", context="talk")

# Generate classification report
report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).transpose()

# Keep only relevant rows (remove accuracy/support summary if needed)
report_df = report_df.iloc[:-1, :]  # removes overall accuracy row

# Select main columns
report_df = report_df[["precision", "recall", "f1-score"]]

# Round values
report_df = report_df.round(2)

# ----------------------------
# Professional Heatmap Style Table
# ----------------------------
plt.figure(figsize=(8, 5))

sns.heatmap(report_df,
            annot=True,
            cmap="Blues",
            cbar=True,
            linewidths=0.5,
            fmt=".2f")

plt.title("Classification Report", fontsize=18, pad=15)
plt.ylabel("Class")
plt.xlabel("Metrics")
plt.tight_layout()
plt.savefig("classification_report_final.png", dpi=300, bbox_inches='tight')
plt.show()
```
